```python
# ----- Input matrix dimensions -----
rows <- as.integer(readline("Enter number of rows: "))
cols <- as.integer(readline("Enter number of columns: "))

# ----- Input elements for Matrix A -----
cat("Enter elements of Matrix A:\n")
A <- matrix(
  as.numeric(scan(n = rows * cols)),
  nrow = rows,
  ncol = cols
)

# ----- Input elements for Matrix B -----
cat("Enter elements of Matrix B:\n")
B <- matrix(
  as.numeric(scan(n = rows * cols)),
  nrow = rows,
  ncol = cols
)

# ----- Display matrices -----
cat("\nMatrix A:\n")
print(A)

cat("\nMatrix B:\n")
print(B)

# ----- Matrix Addition -----
cat("\nMatrix Addition (A + B):\n")
print(A + B)

# ----- Matrix Subtraction -----
cat("\nMatrix Subtraction (A - B):\n")
print(A - B)

# ----- Matrix Multiplication -----
cat("\nMatrix Multiplication (A %*% B):\n")
print(A %*% B)

# ----- Matrix Transpose -----
cat("\nTranspose of Matrix A:\n")
print(t(A))

# ----- Matrix Inverse (if square) -----
if (rows == cols) {
  cat("\nInverse of Matrix A:\n")
  print(solve(A))
} else {
  cat("\nInverse not possible (Matrix is not square)\n")
}

# ----- Matrix Division (A / B â†’ A %*% inverse(B)) -----
if (rows == cols && det(B) != 0) {
  cat("\nMatrix Division (A %*% inverse(B)):\n")
  print(A %*% solve(B))
} else {
  cat("\nMatrix Division not possible (B not invertible)\n")
}
```
