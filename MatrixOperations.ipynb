```python
import numpy as np

# Input matrix dimensions
r = int(input("Enter number of rows: "))
c = int(input("Enter number of columns: "))

# Input Matrix A
print("Enter elements of Matrix A:")
A = np.array([[float(input()) for j in range(c)] for i in range(r)])

# Input Matrix B
print("Enter elements of Matrix B:")
B = np.array([[float(input()) for j in range(c)] for i in range(r)])

# Display matrices
print("\nMatrix A:")
print(A)

print("\nMatrix B:")
print(B)

# Matrix Addition
print("\nMatrix Addition (A + B):")
print(A + B)

# Matrix Subtraction
print("\nMatrix Subtraction (A - B):")
print(A - B)

# Matrix Multiplication
print("\nMatrix Multiplication (A @ B):")
print(A @ B)

# Transpose
print("\nTranspose of Matrix A:")
print(A.T)

# Inverse
if r == c:
    print("\nInverse of Matrix A:")
    print(np.linalg.inv(A))
else:
    print("\nInverse not possible (Matrix not square)")

# Matrix Division
# A / B = A @ inverse(B)
if r == c and np.linalg.det(B) != 0:
    print("\nMatrix Division (A @ inverse(B)):")
    print(A @ np.linalg.inv(B))
else:
    print("\nMatrix Division not possible (B not invertible)")
```
