```python
import pandas as pd

#========CREATE SIMPLE DATA========#
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diane'],
    'Age' : [22,25,23,21],
    'Score' : [85.5,90.0,78.3,92.1]
}

df = pd.DataFrame(data)
print("Original DataFrame:\n", df,"\n")

# ===== EXPORT ==== #
df.to_csv('data.csv', index = False)
df.to_excel('data.xlsx', index = False)
df.to_csv('data.txt', sep='\t',index = False)
print("Files exported: data.csv, data.xlsx,data.txt \n")

# ==== IMPORT ==== #
df_csv = pd.read_csv('/content/data.csv')
df_xls = pd.read_excel('/content/data.xlsx')
df_txt = pd.read_csv('/content/data.txt')

# ==== DISPLAY ==== #
print("--- From CSV ---\n", df_csv,"\n")
print("--- From Excel ---\n", df_xls, "\n")
print("--- From TXT ---\n",df_txt,"\n")
```
